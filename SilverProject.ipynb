```
def userInput(name):
    print(name,"'s Move [R/P/S]: ")
    u_move=input()
    u_move=u_move.upper()
    return u_move
```

```
 
import random
 
def aiInput():
    print("Computer's Move: ")
    moves= ["R", "P", "S"]
    ai_move= random.choice(moves)
    print(ai_move)
    return ai_move
```

```
def Result(user,ai):
 
    up,aip=0,0

    print("\n")
 
    if user=="R" and ai=="R":
        print("It's A Draw")
        Speech(2)
        up=up+1
        aip=aip+1
 
    if user=="R" and ai=="P":
        print("Computer Wins")
        Speech(3)
        aip=aip+2
 
    if user=="R" and ai=="S":
        print("User Wins")
        Speech(1)
        up=up+2
 
    if user=="P" and ai=="R":
        print("User Wins")
        Speech(1)
        up=up+2
 
    if user=="P" and ai=="P":
        print("It's A Draw")
        Speech(2)
        up=up+1
        aip=aip+1
 
    if user=="P" and ai=="S":
        print("Computer Wins")
        Speech(3)
        aip=aip+2
 
    if user=="S" and ai=="R":
        print("Computer Wins")
        Speech(3)
        aip=aip+2
 
    if user=="S" and ai=="P":
        print("User Wins")
        Speech(1)
        up=up+2
 
    if user=="S" and ai=="S":
        print("It's A Draw")
        Speech(2)
        up=up+1
        aip=aip+1
 
    return up,aip
```

```
 
lst= ["'Wow, You Just Beat The Computer!!'", "'Looks Like Someone Just Met It's Match'", "'HAHAHA!! You Thought You Could Beat Me!'"]
 
Speech = lambda num: print(lst[num-1])
```

```
name= input("Enter Your Name:")
 
print("\nLet The Game Begin!!\n")
 
print("[To Exit The Game]\n[Type Anything Other Than R/P/S]")
 
u_move="R" #By Default
 
u_points_sum,ai_points_sum=0,0
 
while u_move=="R" or u_move=="P" or u_move=="S" :
 
    u_move= userInput(name)
    ai_move= aiInput()
 
    
    u_points,ai_points= Result(u_move,ai_move)
 
    u_points_sum = u_points_sum + u_points
    ai_points_sum = ai_points_sum + ai_points 
 
    print("User Got ",u_points," points")
    print("Computer Got ",ai_points," points")
    print("\n\n")
 
print("\n\n\n________Score Card_______")
print("User Got ",u_points_sum," points")
print("Computer Got ",ai_points_sum," points")
print("—————————————————————————")
```
